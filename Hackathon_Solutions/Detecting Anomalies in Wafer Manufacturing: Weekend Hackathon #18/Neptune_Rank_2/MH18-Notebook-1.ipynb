## Load data

```python
import os
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as plt
import seaborn as sns

train = pd.read_csv('./data/Train.csv')
display(train)

test = pd.read_csv('./data/Test.csv')
display(test)
```

## Submission: MH18 Output 03 -- Data Preparation

```python
X = train_dedup.drop(['Class'], axis=1)

y = train_dedup['Class']
```

### Extract outlier indicators as features - LOF, ISOF, KNN, OCSVM

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X2 = train.drop(['Class'], axis=1)

scaler.fit(X2)

X2 = pd.DataFrame(scaler.transform(X2), columns=X2.columns)

train2 = X2.copy()
```

```python
display(train2)
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=50, contamination='auto')

lof.fit_predict(X2)

lof_scores = -lof.negative_outlier_factor_
print(lof_scores)

train2['LOF'] = lof_scores
```

```python
from sklearn.ensemble import IsolationForest

isof = IsolationForest(n_estimators=500,
                       max_features=0.8,
                       bootstrap=True,
                       warm_start=True,
                       random_state=4430)

isof.fit(X2)

isof_scores = -isof.score_samples(X2)
print(isof_scores)

train2['ISOF'] = isof_scores
```

```python
from pyod.models.knn import KNN

knn = KNN(n_neighbors=50)

knn.fit(X2)

knn_scores = knn.decision_scores_

print(knn_scores)
train2['KNN'] = knn_scores
```

```python
from sklearn.svm import OneClassSVM

svm = OneClassSVM(gamma='auto')
svm.fit(X2)

svm_scores = svm.score_samples(X2)

print(svm_scores)

train2['SVM'] = svm_scores
```

```python
svm_mean = train2['SVM'].mean()
train2['SVM'] = train2['SVM'] - svm_mean
```

```python
train2['Class'] = train['Class']

display(train2)
```

```python
train2.to_csv('./data/MH18_Train2.csv', index=False)
```

## Prepare test data to include outlier features

```python
X3 = pd.DataFrame(scaler.transform(test), columns=test.columns)

test2 = X3.copy()

lof.fit_predict(X3)

lof_scores = -lof.negative_outlier_factor_
print(lof_scores)

test2['LOF'] = lof_scores

isof.fit(X3)

isof_scores = -isof.score_samples(X3)
print(isof_scores)

test2['ISOF'] = isof_scores

knn.fit(X3)

knn_scores = knn.decision_scores_

print(knn_scores)
test2['KNN'] = knn_scores

svm.fit(X3)
svm_scores = svm.score_samples(X3)

print(svm_scores)

test2['SVM'] = svm_scores
```

```python
test2['SVM'] = test2['SVM'] - test2['SVM'].mean()

display(test2)
```
