#### Merging weather data into the original dataset

```python
data_weather_agg = data_weather.groupby(by=['LOCATION', 'EVENT','source'], as_index=False)[['AIR_TEMP', 'TRACK_TEMP', 'HUMIDITY', 'PRESSURE',
       'WIND_SPEED', 'WIND_DIRECTION','RAIN']].mean()

data1 = data.merge(data_weather_agg, on = ['LOCATION', 'EVENT'], how='left')

## Filling the null values with mean of values based on Location:
cols = ['AIR_TEMP', 'TRACK_TEMP', 'HUMIDITY', 'PRESSURE',
       'WIND_SPEED', 'WIND_DIRECTION', 'RAIN']

for feat in cols:
    data1[feat] = data1[feat].fillna(data1.groupby(['LOCATION'])[feat].transform('mean'))
```

```python
### Imputing missing values in column S1_LARGE_SEC, S2_LARGE_SEC and S3_LARGE_SEC using DRIVER_NAME_TEAM, LOCATION AND EVENT

### DRIVER_NAME_TEAM : Driver_Name_Team combined

imputed_values0 = data1.loc[~data1['S1'].isna()]
imputed_values0 = imputed_values0.groupby(by=['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], as_index=False)['S1_LARGE_SEC'].mean()
imputed_values0 = imputed_values0.rename(columns={'S1_LARGE_SEC':'S1_imputed'})

imputed_values1 = data1.loc[~data1['S2'].isna()]
imputed_values1 = imputed_values1.groupby(by=['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], as_index=False)['S2_LARGE_SEC'].mean()
imputed_values1 = imputed_values1.rename(columns={'S2_LARGE_SEC':'S2_imputed'})

imputed_values2 = data1.loc[~data1['S3'].isna()]
imputed_values2 = imputed_values2.groupby(by=['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], as_index=False)['S3_LARGE_SEC'].mean()
imputed_values2 = imputed_values2.rename(columns={'S3_LARGE_SEC':'S3_imputed'})

imputed_values = imputed_values0.merge(imputed_values1, on = ['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], how='outer')
imputed_values = imputed_values.merge(imputed_values2, on = ['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], how='outer')
data2 = data1.merge(imputed_values, on = ['DRIVER_NAME_TEAM','LOCATION', 'EVENT'], how='left')
```

```python
data2.loc[data2['S1'].isna(), 'S1_LARGE_SEC'] = data2['S1_imputed']
data2.loc[data2['S2'].isna(), 'S2_LARGE_SEC'] = data2['S2_imputed']
data2.loc[data2['S3'].isna(), 'S3_LARGE_SEC'] = data2['S3_imputed']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, random_state=12, test_size=0.15)
```

#### Extreme Gradient Boost Regressor

```python
### Tuning XGB

param_grid = {
    'learning_rate' : [0.5]
  , 'n_estimators' : [150]
  , 'max_depth': [8]
  , 'min_child_weight': [1]
  , 'subsample': [1]
  , 'colsample_bytree': [1]
  , "reg_lambda" :  [0.2]
  , "reg_alpha" : [0]
}

xgb_reg = xgb.XGBRegressor(objective='reg:squaredlogerror', 
                           booster='gbtree',validate_parameters=True, random_state = 69, n_jobs=-1) 

### fitting the model
model = GridSearchCV(estimator=xgb_reg,param_grid=param_grid,scoring='neg_mean_squared_log_error',
                    cv=5, return_train_score=True, n_jobs=-1, verbose=10)
model.fit(X_train, Y_train)
```

```python
####### checking the results of the validation
summary=pd.DataFrame(model.cv_results_)

display=summary[['rank_test_score','param_learning_rate', 'param_max_depth',
       'param_min_child_weight', 'param_n_estimators', 'param_reg_alpha',
       'param_reg_lambda','split0_test_score', 'split1_test_score', 'split2_test_score',
       'split3_test_score','split4_test_score', 'mean_test_score',
       'std_test_score', 'split0_train_score',
       'split1_train_score', 'split2_train_score', 'split3_train_score', 'split4_train_score', 'mean_train_score', 'std_train_score',
       
                ]]
display.sort_values(by='rank_test_score')
```

```python
## testing out on X_val:
Y_val_pred = model.predict(X_val)
print(np.sqrt(mean_squared_log_error(Y_val, Y_val_pred)))
Y_train_pred = model.predict(X_train)
print(np.sqrt(mean_squared_log_error(Y_train, Y_train_pred)))
```
